This notebook creates a linear regression model in python to model salary based on years of experience from a csv file containing numerical data.

## Import necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
```

```python
# import csv file into pandas df
df = pd.read_csv("regression_data.csv")

# make a scatter plot of Salary (y) as a fx of YearsExperience (x)
plt.scatter(df['YearsExperience'], df['Salary'], color='purple')
```

## Fit linear regression model 
This code will fit a linear regression model using the scipy linregress command to the imported data. It will model salary as a fx of years of experience.


```python
x = df[['YearsExperience']]
y = df[['Salary']]

slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = slope * x + intercept
mse = mean_squared_error(y, y_pred)


print(f'Slope for linear regression: {slope[0]:.3f}')
print(f'Intercept for linear regression: {intercept[0]:.3f}')
# print(f'Standard error for linear regression: {std_err[0]:.3f}')
print(f'R^2 value for linear regression: {r_value[0]:.3f}')
print(f'MSE for linear regression: {mse:.3f}')

type(mse)
```

```python
# graph the raw data and the calculated regression line
plt.scatter(df['YearsExperience'], df['Salary'], color='purple') # graph data from original table
plt.plot(df['YearsExperience'], y_pred, 'r-', label = "Fitted Line") # add in calculated linear regression using model previously generated
plt.text(1.5, df['Salary'].max() - 5000,
         f"y = {slope[0]:.2f}x + {intercept[0]:.2f}\n"
         f"r = {r_value[0]:.2f}\nMSE = {mse:.2f}",
         fontsize=12)
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.savefig('linear_regression_python.png')
plt.show()
```
