create a linear regression model (python) to model salary based on years of experience

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# import csv file into pandas df
df = pd.read_csv("regression_data.csv")

# print(df)

# make a scatter plot of Salary (y) as a fx of YearsExperience (x)
plt.scatter(df['YearsExperience'], df['Salary'], color='purple')
```

```python
# fit a linear regression model to the data (salary as a fx of years of experience)
model = LinearRegression().fit(df[['YearsExperience']], df[['Salary']])
```

```python
# graph the raw data and the calculated regression line
plt.scatter(df['YearsExperience'], df['Salary'], color='purple') # graph data from original table
plt.plot(df['YearsExperience'], model.predict(df[['YearsExperience']]), color='green') # add in calculated linear regression using model previously generated
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# evaluate the model's performance
r = model.score(df[['YearsExperience']] ,df[['Salary']]) # calculate the R squared (fit)

print(f'R^2 value for linear regression: {r}')
```

```python
# graph the data and linear regression using seaborn regplot (shading shows 95% CI) 
sns.set_style('white')
sns.set_palette('Set2')
ax = sns.regplot(x ='YearsExperience', y ='Salary', data = df)
ax.set_title('Salary vs. Experience')
```
